# Simple Chatbot using predefined responses

```python
# Import the regular expression module to handle pattern matching
import re
```

```python
# A dictionary that maps keywords to predefined responses
responses = {
    "hello": "Hi there! How can I assist you today?",
    "hi": "Hello! How can I help you?",
    "how are you": "I'm just a bot, but I'm doing great! How about you?",
    "what is your name": "I'm a chatbot created to assist you.",
    "help": "Sure, I'm here to help. What do you need assistance with?",
    "thank you": "You're welcome! I'm happy to help.",
    "default": "I'm not sure I understand. Could you please rephrase?"
}
```

```python
# Function to find appropriate respone based on the user's input
def chatbot_response(user_input):
  # Convert user input to lowercase to make matching case-insensitive
  user_input = user_input.lower()

  for keywords in responses:
    if re.search(keywords, user_input):
      return responses[keywords]

  return responses["default"]
```

```python
# Main funtion to run the chatbot
def chatbot():
  print("Chatbot: Hello! I'm a simple chatbot, i'm here to assist you. (Type 'bye' to exit)")
  while True:
    user_input = input("You: ")

    # if user type bye, exist the loop
    if user_input.lower() == "bye":
      print("Chatbot: Goodbye! Have a nice day!")
      break
    else:
      # Get chatbot's respone based on input user
      response = chatbot_response(user_input)
      # Print chatbot's respone
      print(f"Chatbot: {response}")
```

```python
# Run chatbot
chatbot()
```
